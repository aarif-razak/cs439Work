```python
#Transform functions may run across columns
#display(df)
df_grouped = df.groupby('Product Type')
display(df_grouped.transform(func= np.max)) #for each group, will give us the max value per column
```

```python
#filter function, filter our entries by a certain value, like having a desktop or not.
#Filter function MUST return a boolean value.

def filter_f(group):
    return len(group) >= 15

df_grouped.filter(func=filter_f)
```

```python
#apply function
df_grouped.apply(func=np.mean) #apply similar to transform. Transform works column wise, and returns a series/df (cant change output size)
#apply works on teh whole dataframe and can change the size of the output
```
