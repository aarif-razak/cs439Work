```python
df['State'].value_counts()
```

```python
ax = df['State'].value_counts()[['OH', 'PA', 'NY', 'WV']].plot(kind= 'bar') #you can reorder on columsn of choice
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
ax.set_xlabel('State')
ax.set_ylabel('Num of sales') #y tick labels are automatically filled
ax.set_title('Number of sales per state')
```

```python
#now we can use seaborn's plotting
import seaborn as sns
```

```python
ax = sns.countplot(data = df, x = 'State', order = ['OH', 'PA', 'NY', 'WV'])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
ax.set_xlabel('State')
ax.set_ylabel('Num of sales') #y tick labels are automatically filled
ax.set_title('Number of sales per state')
```

```python
ax = sns.boxplot(data = df,  x = 'State', y = 'Profit')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
ax.set_xlabel('State')
ax.set_ylabel('Num of sales') #y tick labels are automatically filled
ax.set_title('Number of sales per state')
```
