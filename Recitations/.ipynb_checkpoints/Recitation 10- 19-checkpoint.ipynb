```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
import seaborn as sns
import scipy

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,9)
```

```python
path = 'data/lion.jpeg'
img = mpimg.imread(path)
imgplot = plt.imshow(img)
```

```python
img.shape
```

```python
dimensions = img.shape
R = np.zeros((dimensions[0], dimensions[1]),int)
G = np.zeros((dimensions[0], dimensions[1]),int)
B = np.zeros((dimensions[0], dimensions[1]),int)

R1 = img[:,:,0]
R1
for i in range(dimensions[0]):
    R[i,:] = img[i,:,0]
    G[i,:] = img[i,:,1]
    B[i,:] = img[i,:,2]
    
print(R1 == R)
```

```python
#rank 1 approx
from scipy import linalg
U0, s0, v0 = linalg.svd(R)
U1, s1, v1 = linalg.svd(G)
U2, s2, v2 = linalg.svd(B)

print(f'{U0.shape}, {s0.shape}, {v0.shape}')
print(s0,s1,s2)
```

```python
#rank -1 approx
rank1_approx_R = U0[:,0].reshape(len(U0[:,0]),1) *s0[0] *v0[0,:]
rank1_approx_G = U1[:,0].reshape(len(U0[:,0]),1) *s1[0] *v1[0,:]
rank1_approx_B = U2[:,0].reshape(len(U0[:,0]),1) *s2[0] *v2[0,:]

rank1_approx = np.zeros(dimensions, int)
```

```python
for i in range(dimensions[0]):
    rank1_approx[i,:,0] = rank1_approx_R[i,:]
    rank1_approx[i,:,1] = rank1_approx_G[i,:]
    rank1_approx[i,:,2] = rank1_approx_B[i,:]

plt.imshow(rank1_approx)
```

```python
#rank -1 approx
rank1_approx_R = U0[:,0].reshape(len(U0[:,0]),1) *s0[0] *v0[0,:]
rank1_approx_G = U1[:,0].reshape(len(U0[:,0]),1) *s1[0] *v1[0,:]
rank1_approx_B = U2[:,0].reshape(len(U0[:,0]),1) *s2[0] *v2[0,:]
#rank 2
rank2_approx_R = U0[:,1].reshape(len(U0[:,0]),1) *s0[1] *v0[1,:] #this changes from 0 to 1
rank2_approx_G = U1[:,1].reshape(len(U0[:,0]),1) *s1[1] *v1[1,:]
rank2_approx_B = U2[:,1].reshape(len(U0[:,0]),1) *s2[1] *v2[1,:]

rank1_approx = np.zeros(dimensions, int)
rank2_approx = np.zeros(dimensions, int)


for i in range(dimensions[0]):
    rank2_approx[i,:,0] = rank1_approx_R[i,:] + rank2_approx_R[i,:] #following summation formula, adding the other approx.
    rank2_approx[i,:,1] = rank1_approx_G[i,:] + rank2_approx_G[i,:]
    rank2_approx[i,:,2] = rank1_approx_B[i,:] + rank2_approx_B[i,:]

plt.imshow(rank2_approx)
```
