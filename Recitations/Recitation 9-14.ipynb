```python
import numpy as np
import pandas as pd
```

```python
#create a series
a = pd.Series(data = np.linspace(10,100, num= 10))
print(a)
```

```python
index = [ chr(ord('a')+i) for i in range(10)]
#Make a list, use ord to take a string character, and return the cooresponding integer value, within a for loop 

print(index)
```

```python
# create a seeis with this index
a = pd.Series(data = np.linspace(10,100, num= 10), index = index, name = 'numbers')
#name is used to set the column name of the series
print(a)
```

```python
print(c)
```

```python
#operations on series
a = pd.Series(data = np.linspace(10, 100, 10))
b = pd.Series(data = np.linspace(30, 300, 10))

c = a+b #number of elements in both series should be the same
# each of the series (a,b) have the same 10 number of elements

print(c)

print(c.max()) #built in methods within a Series (like max)
```

```python
#slicing within a Series
print(a[3:8]) #non inclusive (prints from 3 to 7)
print(a.loc[3]) #use the index of 3 to print it out (technically the fourth element)
```

```python
print(a.loc[2:4]) #INCLUSIVE OF FOUR
```

```python
print(a.iloc[2:4]) #NON INCLUSIVE OF THE LAST ELEMENT
```

```python
a1 = a.drop(index = [0,1]) #remove the indexes from 0 TO (and including) 1
#will create a new copy of the array
print(a) #these are two seperate arrays!
print(a1) #use the 'inplace' argument within a.drop to avoid the expensive cost of remaking the array, since you can directly manipulate the array
```
