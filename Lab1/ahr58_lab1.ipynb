### Linearly Spaced Points
There are many tasks that you need to generate a set of points in a range that are linearly spaced. For this task
we will use numpy modules arnage and linspace.

Note that the tutorial uses `np.arange`. While this is fine in some cases, we generally prefer to use `np.linspace`. `np.linspace(a, b, N)` divides the interval `[a, b]` into N points, while `np.arange(a, b, s)` will step from `a` to `b` with a fixed step size `s`.

One thing to keep in mind is that `np.linspace` always includes both end points while `np.arange` will *not* include the second end point `b`. For this reason, when we are plotting ranges of values we tend to prefer `np.linspace`.

Write two functions that return the same result.

```python
# create an array of linearly spaced points from -5(inclusive) to 6 (not inclusive) spaced by 1
A = np.arange(-5,6)
print(A)

# do the same with linspace
B = np.linspace(-5,5, 11) #11 is our N value since this will space values by 1
print(B)
```

### Activity 2
Change the function to $$f(x,y) = xy $$ and re-plot. Estimate the values (x,y) that minimizes f(x,y)?

<h1>The values of (x,y) that could lead towards a minimum of the function f(t) could be as y approaches -∞ and as x approaches positive ∞ </h1>

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x*y)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

## Task 3 - Reading Documentation
We should also familiarize ourselves with looking up documentation and learning how to read it. Below is a section of code that plots a basic wireframe. Replace each `# Your answer here` with a description of what the line above does, what the arguments being passed in are, and how the arguments are used in the function. For example your answer may look like,

`np.arange(2, 5, 0.2)`

`# This returns an array of numbers from 2 to 5 with an interval size of 0.2`

Hint: The `Shift` + `Tab` (to see what a function does) tip from earlier in the notebook may help here.

```python
from mpl_toolkits.mplot3d import axes3d

u = np.linspace(1.5*np.pi, -1.5*np.pi, 100)
# Returns evenly spaced numbers over a specified interval.
# In this case, the 'start', 'stop', and 'num' arguments mean that we will have 50 values in the interval of 1.5*pi -> -1.5*pi
[x,y] = np.meshgrid(u, u)
# Takes in 1-D coordinate vectors and returns N dimention coordinate matricies (these are placed in a list in this scenario). 
# The linspace 'u' represents the coordinates of a grid and the indexing is done via a default Cartesian coordinate plane.
squared = np.sqrt(x.flatten()**2 + y.flatten()**2)
z = np.cos(squared)
# This returns an ndarray for cooresponding cosine values on the inputted array in radians. In this case, this is the modified 'squared' variable.
z = z.reshape(x.shape)
# Reshapes the ndarray 'z' into that of the shape of 'x'. This will return the same data, but a different shape.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# This will produce a sublpot specified with 1 row(s), 1 column(s), and at the index of 1. The 'projection' alternate argument will produce a 3d subplot (in this case.)
# Normally, the inputs of columns, rows, and index are seperate integers but can be substituted with a single integer. 
# Example: fig.add_subplot(235) is the same as fig.add_subplot(2, 3, 5).
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
# This will plot the wireframe with 2d arrays of 'x', 'y', and 'z' with a downsampling stride of 10 in each direction.
ax.view_init(elev=50., azim=30)
# This will set the elevation and azimuth of the axes in degrees. 
# In this case, the elevation angle is 50 and the azimtuh angle is 30 in the (x,y) plane.
plt.savefig("figure1.png")
```
